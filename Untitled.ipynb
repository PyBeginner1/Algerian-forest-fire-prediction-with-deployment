```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models = {
    'Logistic' : LogisticRegression(),
    'Decision Tree' : DecisionTreeClassifier(),
    'SVM' : SVC(),
    'RandomForest': RandomForestClassifier(),
    'Naive Bayes' : GaussianNB()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + ' trained!')
```

```python
acc = []
for name, model in models.items():
    y_pred = model.predict(X_test)
    acc.append(round(accuracy_score(y_test, y_pred)*100, 2))
    score = round(accuracy_score(y_test, y_pred)*100, 2)
    print(name , ' accuracy is: ' , score,'%')
```

```python
import pickle as pkl
```

```python
tree_regressor = DecisionTreeClassifier()
tree_regressor.fit(X_train, y_train)
```

```python
pkl.dump(tree_regressor, open('decision_tree.pkl', 'wb'))
```

```python
pickled_model = pkl.load(open('decision_tree.pkl', 'rb'))
pickled_model.predict(X_test)
```
